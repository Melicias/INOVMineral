```python
#data.dropna(thresh=1)
data['history'].unique()
```

```python
def count_upper_string(string):
    upper = 0
    for c in string:
      if not c.islower():
        upper += 1
    return upper

def count_lower_string(string):
    lower = 0
    for c in string:
      if c.islower():
          lower += 1
    return lower

data['history_originator'] = data.apply(lambda row: count_upper_string(row['history']), axis=1)
data['history_responder'] = data.apply(lambda row: count_lower_string(row['history']), axis=1)
```
